# Python Basics for Underwriting Decisions

This notebook introduces core Python concepts (variables, data types, and logic)
using a simplified insurance underwriting example.


```python
# Applicant information
age = 42
is_smoker = True
annual_income = 650000  # ZAR
```

```python
type(age), type(is_smoker), type(annual_income)
```

```python
if age > 40 and is_smoker:
    risk_decision = "High Risk"
else:
    risk_decision = "Standard Risk"

risk_decision
```

## Underwriting Interpretation

- Applicants over age 40 who smoke are classified as **High Risk**
- All other applicants are treated as **Standard Risk**

This logic mirrors a simplified underwriting rule translated into executable code.


```python
# Second applicant
age_2 = 28
is_smoker_2 = False
annual_income_2 = 420000
```

```python
if age_2 > 40 and is_smoker_2:
    risk_decision_2 = "High Risk"
else:
    risk_decision_2 = "Standard Risk"

risk_decision_2
```

## Comparing Applicants

- Applicant 1 (42, smoker) → **High Risk**
- Applicant 2 (28, non-smoker) → **Standard Risk**

The same rule produces different outcomes based on applicant characteristics.


```python
if age > 40 and is_smoker and annual_income < 800000:
    risk_decision = "High Risk"
else:
    risk_decision = "Standard Risk"

risk_decision
```

## Reflection

This notebook shows how underwriting logic can be represented as:
- Variables (applicant attributes)
- Conditions (risk rules)
- Outputs (risk classification)

Even simple Python can model real underwriting decision-making.
